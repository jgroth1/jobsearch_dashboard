```python
import json
import os
from bs4 import BeautifulSoup
import re
from nltk.stem import WordNetLemmatizer
from nltk import sent_tokenize, wordpunct_tokenize, pos_tag
from nltk.corpus import stopwords
from nltk.util import ngrams
import nltk
from gensim.models.doc2vec import TaggedDocument, Doc2Vec
from gensim import corpora, models, similarities
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def paras(html):
    soup = BeautifulSoup(html, 'html.parser')
    for string in soup.stripped_strings:
        yield string

def sents(html):
    for paragraph in paras(html):
        for sentence in sent_tokenize(paragraph):
            yield sentence

def words(html):
    for sentence in sents(html):
        for token in wordpunct_tokenize(sentence):
            yield token

def tokenize(html):
    for paragraph in paras(html):
        yield [pos_tag(wordpunct_tokenize(sent)) for sent in sent_tokenize(paragraph)]

def ngramizer(html, n, stop_words):
    wnl = WordNetLemmatizer()
    
    for sent in sents(html):
        sent_list = sent.split()
        lemma_sent = []
        for x in sent_list:
            x = x.lower()
            if x not in stop_words:
                lemma_sent.append(wnl.lemmatize(str(x)))
        
        for ngram in ngrams(lemma_sent, n, pad_left=True, pad_right=True, left_pad_symbol='<s>', right_pad_symbol='</s>'):
            yield ngram
    
```

```python
docs = []
titles = []
for i in range(10):
    i += 1
    path = 'corpus3/02-09-2018/Page_' + str(i) + '/'
    files = os.listdir(path)
    rxs = [re.search('^\.',file) for file in files]
    i =[n for n in range(len(rxs)) if rxs[n]]
    n = 0
    for k in i:
        files.pop(k-n)
        n += 1
    n=0
    for file in files: 
        
        with open(path + file, encoding='utf-8') as f:
            job_post = f.read()

        job_dict = eval(job_post)

        job_dict.keys()

        title = job_dict['job title']
        company = job_dict['company']
        titles.append([n, title, company])
        html = job_dict['job description']

        stop_words = set(stopwords.words('english'))
        wnl = WordNetLemmatizer()

        doc_gram = [x for x in ngramizer(html, 1, stop_words)]
        docs.append(doc_gram)
        n += 1
```
