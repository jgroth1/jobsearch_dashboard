```python
import RAKE
from bs4 import BeautifulSoup
import os
import re
import pandas as pd
```

```python
keywords = []
titles = []
for i in range(1):
    i += 1
    #path = 'corpus3/02-09-2018/Page_' + str(i) + '/'
    path = 'Careeer_project11-09-2018/Page_' + str(i) + '/'
    files = os.listdir(path)
    rxs = [re.search('^\.',file) for file in files]
    i =[n for n in range(len(rxs)) if rxs[n]]
    n = 0
    for k in i:
        files.pop(k-n)
        n += 1
    n=0
    doc = []
    for file in files: 
        
        with open(path + file, encoding='utf-8') as f:
            job_post = f.read()

        job_dict = eval(job_post)

        job_dict.keys()

        title = job_dict['job title']
        company = job_dict['company']
        titles.append([n, title, company])
        html = job_dict['job description']
        soup = BeautifulSoup(html, 'html.parser')
        stripped = soup.stripped_strings
        
        for string in stripped:
            doc.append(string)
        flat_doc = [item for item in doc]
        joined_doc = ' '.join(flat_doc)
        rake3 = RAKE.Rake('stop.txt')
        keys = rake3.run(joined_doc, minCharacters = 1, maxWords = 2, minFrequency = 1)
        keywords.append(keys)
print(keywords)
```

```python
flat_keys = [item for items in keywords for item in items]
print(flat_keys)
```

```python
keys_df = pd.DataFrame(flat_keys)
keys_df.head()
```

```python
keys_grouped = keys_df.groupby(0).count()
keys_grouped.sort_values(1, ascending=False, inplace=True)
```

```python
keys_df.to_csv('buncha_keyswords2.csv')
```
