```python
import json
import os
from bs4 import BeautifulSoup
import re
from nltk import sent_tokenize, wordpunct_tokenize, pos_tag
import nltk
from gensim.models.doc2vec import TaggedDocument, Doc2Vec
from gensim import corpora, models, similarities
import nltk, math, codecs
from nltk.cluster.kmeans import KMeansClusterer
```

```python
path = 'corpus2/31-08-2018/Page_10/'
files = os.listdir(path)
rxs = [re.search('^\.',file) for file in files]
i =[n for n in range(len(rxs)) if rxs[n]]
n = 0
for k in i:
    files.pop(k-n)
    n += 1

def paras(html):
    soup = BeautifulSoup(html, 'html.parser')
    for string in soup.stripped_strings:
        yield string

def sents(html):
    for paragraph in paras(html):
        for sentence in sent_tokenize(paragraph):
            yield sentence

def words(html):
    for sentence in sents(html):
        for token in wordpunct_tokenize(sentence):
            yield token

def tokenize(html):
    for paragraph in paras(html):
        yield [pos_tag(wordpunct_tokenize(sent)) for sent in sent_tokenize(paragraph)]
    

docs = []
titles = []
n=0
for file in files: 
    with open(path + file, encoding='utf-8') as f:
        job_post = f.read()

    job_dict = eval(job_post)

    job_dict.keys()

    title = job_dict['job title']
    company = job_dict['company']
    titles.append([n, title, company])
    html = job_dict['job description']

    
    doc = [x for x in words(html)]
    docs.append(doc)
    n += 1
```

```python
path = 'corpus2/31-08-2018/Page_25/'
files = os.listdir(path)
rxs = [re.search('^\.',file) for file in files]
i =[n for n in range(len(rxs)) if rxs[n]]
n=0
for k in i:
    files.pop(k-n)
    n += 1
n=0
for file in files: 
    with open(path + file, encoding='utf-8') as f:
        job_post = f.read()

    job_dict = eval(job_post)

    job_dict.keys()

    title = job_dict['job title']
    company = job_dict['company']
    titles.append([n, title, company])
    html = job_dict['job description']

    
    doc = [x for x in words(html)]
    docs.append(doc)
    n += 1
    
```

```python
print(files)
```

```python
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(docs)]
model = Doc2Vec(documents, vector_size=50, min_count=3)
vector = []
for doc in docs:
    inferred_vector = model.infer_vector(doc)
    vector.append(inferred_vector)
```

```python
NUM_CLUSTERS = 10
kclusterer = KMeansClusterer(NUM_CLUSTERS, distance=nltk.cluster.util.cosine_distance, repeats=25)
assigned_clusters = kclusterer.cluster(vector, assign_clusters=True)
```

```python
print(assigned_clusters)
```

```python
cg = []
for i in range(NUM_CLUSTERS):
    l = []
    for j in range(len(docs)):
        if assigned_clusters[j] == i:
            l.append(docs[i])
    cg.append(l)
```
