```python
from splinter import Browser
from bs4 import BeautifulSoup
import datetime
import os
import json
import time
import re
```

```python
url = 'https://www.glassdoor.com/Job/data-scientist-jobs-SRCH_KO0,14_IP2.htm'

executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)

browser.visit(url)
time.sleep(2)
x_button = browser.find_by_css('div.xBtn')
if x_button:
    x_button.click()
site_list = browser.find_by_xpath("//script[@type='application/ld+json']")
scripts = browser.find_by_tag('script')
print(site_list)
```

```python
for script in scripts:
    print('##################################################################################')
    print(script.html)
    print('##################################################################################')
```
